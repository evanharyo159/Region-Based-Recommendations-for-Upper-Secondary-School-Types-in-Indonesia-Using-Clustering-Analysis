```python
import numpy as np
import pandas as pd
from sklearn import cluster
from pyclustering.cluster.kmedoids import kmedoids
from sklearn.metrics import pairwise_distances
from scipy.spatial import distance
from itertools import combinations

X_np = X.to_numpy()  # Convert DataFrame to NumPy array
# Define a function to calculate the Dunn Index
def dunn_index(clusters, distance_matrix):
    max_intracluster_distances = []
    min_intercluster_distances = []

    # Calculate maximum intra-cluster distances
    for cluster in clusters:
        distances = distance_matrix[np.ix_(cluster, cluster)]
        max_intracluster_distances.append(np.max(distances))

    # Calculate minimum inter-cluster distances
    cluster_combinations = list(combinations(clusters, 2))
    for combo in cluster_combinations:
        distances = distance_matrix[np.ix_(combo[0], combo[1])]
        min_intercluster_distances.append(np.min(distances))

    # Calculate Dunn Index
    dunn_index = min(min_intercluster_distances) / max(max_intracluster_distances)
    return dunn_index

# Create instances of clustering algorithms
clustering_algorithms = {
    "K-Means 4": cluster.KMeans(n_clusters=4, init='random',random_state = 99),
    "K-Means 5": cluster.KMeans(n_clusters=5, init='random',random_state = 99),
    "K-Means 6": cluster.KMeans(n_clusters=6, init='random',random_state = 99),

}

# Calculate Dunn Index for each clustering algorithm
dunn_scores = {}
for name, algorithm in clustering_algorithms.items():
    algorithm.fit(X)
    labels = algorithm.labels_
    clusters = [np.where(labels == i)[0] for i in range(max(labels) + 1)]
    distance_matrix = pairwise_distances(X)
    dunn = dunn_index(clusters, distance_matrix)
    dunn_scores[name] = dunn

# Print Dunn Index scores for comparison
for name, dunn in dunn_scores.items():
    print(f"{name}: Dunn Index = {dunn:.4f}")
```

```python
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from sklearn.metrics import pairwise_distances

# Function to calculate Dunn Index
def dunn_index(X, clusters):
    # Compute all pairwise distances between points in X
    distances = pairwise_distances(X)

    # Initialize variables to track minimum inter-cluster distance and maximum intra-cluster distance
    min_inter_cluster_dist = float('inf')
    max_intra_cluster_dist = float('-inf')

    # Loop through all clusters to calculate intra-cluster and inter-cluster distances
    for cluster_idx, cluster in enumerate(clusters):
        # Get all points in the current cluster
        cluster_points = X[cluster]

        # Calculate the maximum distance within the cluster (intra-cluster distance)
        intra_cluster_dist = np.max(pairwise_distances(cluster_points))
        max_intra_cluster_dist = max(max_intra_cluster_dist, intra_cluster_dist)

        # Calculate the minimum distance between this cluster and all other clusters (inter-cluster distance)
        for other_cluster_idx, other_cluster in enumerate(clusters):
            if cluster_idx != other_cluster_idx:
                # Get all points in the other cluster
                other_cluster_points = X[other_cluster]

                # Compute the distance between all points in the two clusters
                inter_cluster_dist = np.min(pairwise_distances(cluster_points, other_cluster_points))
                min_inter_cluster_dist = min(min_inter_cluster_dist, inter_cluster_dist)

    # Dunn Index formula
    return min_inter_cluster_dist / max_intra_cluster_dist

# Specify the number of clusters (K) for each case
k_values = [4, 5, 6]  # We will calculate the Dunn Index for 4, 5, and 6 clusters
np.random.seed(99)

# Sample size for Clara
sample_size = 514  # Adjust this based on your dataset size
iterations = 20  # Number of times to run the algorithm with different samples

# Create a KMedoids instance and the data (e.g., X is your dataset)
X_np = X.to_numpy()  # Convert DataFrame to NumPy array

# Store Dunn Index for each cluster count
dunn_indexes = {}

# Clara Algorithm (multiple random initializations)
for k in k_values:
    dunn_indexes[k] = []

    for i in range(iterations):
        # Randomly sample a subset of the data
        sample_indices = np.random.choice(len(X_np), sample_size, replace=False)
        X_sample = X_np[sample_indices]

        # Perform K-Medoids on the sampled data
        kmedoids_instance = kmedoids(X_sample, initial_index_medoids=np.random.choice(len(X_sample), k, replace=False))
        kmedoids_instance.process()

        # Get the clusters
        clusters = kmedoids_instance.get_clusters()

        # Calculate Dunn Index for the current clustering
        dunn = dunn_index(X_sample, clusters)
        dunn_indexes[k].append(dunn)

# Print the Dunn Index for each cluster size
for k, scores in dunn_indexes.items():
    avg_dunn_index = np.mean(scores)
    print(f"CLARA {k}: Dunn Index = {avg_dunn_index:.4f}")
```

```python
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from sklearn.metrics import davies_bouldin_score

# Function to calculate Davies-Bouldin Index for a clustering
def davies_bouldin_index(X, clusters):
    # Flatten the cluster labels
    labels = np.zeros(len(X))
    for cluster_idx, cluster in enumerate(clusters):
        labels[cluster] = cluster_idx

    # Calculate the Davies-Bouldin Index
    return davies_bouldin_score(X, labels)

# Specify the number of clusters (K) for each case
k_values = [4, 5, 6]  # We will calculate the Davies-Bouldin Index for 4, 5, and 6 clusters
np.random.seed(99)

# Sample size for Clara
sample_size = 514  # Adjust this based on your dataset size
iterations = 20  # Number of times to run the algorithm with different samples

# Create a KMedoids instance and the data (e.g., X is your dataset)
X_np = X.to_numpy()  # Convert DataFrame to NumPy array

# Store Davies-Bouldin Index for each cluster count
db_indexes = {}

# Clara Algorithm (multiple random initializations)
for k in k_values:
    db_indexes[k] = []

    for i in range(iterations):
        # Randomly sample a subset of the data
        sample_indices = np.random.choice(len(X_np), sample_size, replace=False)
        X_sample = X_np[sample_indices]

        # Perform K-Medoids on the sampled data
        kmedoids_instance = kmedoids(X_sample, initial_index_medoids=np.random.choice(len(X_sample), k, replace=False))
        kmedoids_instance.process()

        # Get the clusters
        clusters = kmedoids_instance.get_clusters()

        # Calculate Davies-Bouldin Index for the current clustering
        db_index = davies_bouldin_index(X_sample, clusters)
        db_indexes[k].append(db_index)

# Print the Davies-Bouldin Index for each cluster size
for k, scores in db_indexes.items():
    avg_db_index = np.mean(scores)
    print(f"CLARA {k}: Davies-Bouldin Index = {avg_db_index:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn import cluster
from sklearn.metrics import silhouette_score

# Assuming you have loaded or generated your data as X
df = pd.read_excel('All_Indonesia_Focus_Clustering_edit.xlsx', sheet_name='edit')
df.head()
X = df.iloc[:,2:]
X
```

```python
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
import matplotlib.pyplot as plt

# Specify the number of clusters (K)
k = 6
np.random.seed(99)

# Sample size for Clara
sample_size = 514  # Adjust this based on your dataset size
iterations = 20  # Number of times to run the algorithm with different samples

# Create a KMedoids instance and the data (e.g., X is your dataset)
X_np = X.to_numpy()  # Convert DataFrame to NumPy array

# Store the best clustering result
best_cost = float('inf')
best_iteration = None
best_clusters = None
best_medoids = None

# Clara Algorithm (multiple random initializations)
for i in range(iterations):
    # Randomly sample a subset of the data
    sample_indices = np.random.choice(len(X_np), sample_size, replace=False)
    X_sample = X_np[sample_indices]

    # Perform K-Medoids on the sampled data
    kmedoids_instance = kmedoids(X_sample, initial_index_medoids=np.random.choice(len(X_sample), k, replace=False))
    kmedoids_instance.process()

    # Get the clusters and medoids
    clusters = kmedoids_instance.get_clusters()
    medoids = kmedoids_instance.get_medoids()

    # Calculate the total cost (sum of dissimilarities from medoids)
    total_cost = sum([np.sum([np.linalg.norm(X_sample[i] - X_sample[medoid]) for i in cluster])
                      for cluster, medoid in zip(clusters, medoids)])

    # Check if this result is the best
    if total_cost < best_cost:
        best_cost = total_cost
        best_iteration = i
        best_clusters = clusters
        best_medoids = medoids

# Print the best iteration and its cost
print(f"Best Iteration: {best_iteration + 1}")
print(f"Best Cost: {best_cost}")

# After finding the best clustering, assign the cluster labels to the full dataset
df['Clara Cluster'] = None
for i, cluster in enumerate(best_clusters):
    for index in cluster:
        df.loc[df.index[index], 'Clara Cluster'] = i + 1  # Assign cluster label

# Print the total number of cities in each cluster
cluster_counts = df['Clara Cluster'].value_counts()
print("\nTotal number of cities in each cluster:")
print(cluster_counts)

# Plotting the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X_np[:, 0], X_np[:, 1], c=df['Clara Cluster'], cmap='viridis', s=50, alpha=0.6)

# Visualize the medoids of the best clustering
medoid_coordinates = X_np[best_medoids]

# Plot the cluster centers (medoids) in red
plt.scatter(medoid_coordinates[:, 0], medoid_coordinates[:, 1], s=200, c='red', marker='X', label='Medoids')

plt.title(f'Clara Clustering (Best Iteration: {best_iteration + 1})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from pyclustering.cluster.kmedoids import kmedoids
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set parameters
k = 6
np.random.seed(99)
sample_size = 514
iterations = 20
X_np = X.to_numpy()

# Initialize best result tracking
best_cost = float('inf')
best_iteration = None
best_clusters = None
best_medoids = None

# Run CLARA
for i in range(iterations):
    sample_indices = np.random.choice(len(X_np), sample_size, replace=False)
    X_sample = X_np[sample_indices]
    initial_medoids = np.random.choice(len(X_sample), k, replace=False)
    kmedoids_instance = kmedoids(X_sample, initial_medoids)
    kmedoids_instance.process()
    clusters = kmedoids_instance.get_clusters()
    medoids = kmedoids_instance.get_medoids()

    # Calculate cost
    total_cost = sum([np.sum([np.linalg.norm(X_sample[i] - X_sample[medoid]) for i in cluster])
                      for cluster, medoid in zip(clusters, medoids)])

    if total_cost < best_cost:
        best_cost = total_cost
        best_iteration = i
        best_clusters = clusters
        best_medoids = medoids

# Assign cluster labels
df['Clara Cluster'] = None
for cluster_idx, cluster in enumerate(best_clusters):
    for index in cluster:
        df.loc[df.index[index], 'Clara Cluster'] = cluster_idx

# PCA for 3D Visualization
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_np)
pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2', 'PCA3'])
pca_df['Cluster'] = df['Clara Cluster']

# 3D Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    pca_df['PCA1'], pca_df['PCA2'], pca_df['PCA3'],
    c=pca_df['Cluster'], cmap='Set1', s=80, alpha=0.8
)

ax.set_title(f'3D PCA Visualization of CLARA Clustering (Best Iteration: {best_iteration + 1})')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

plt.show()
```
